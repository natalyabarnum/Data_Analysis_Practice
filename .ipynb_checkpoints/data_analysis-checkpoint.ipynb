# Let's begin the analysis

### What industry do developers get paid the most?

```python
stack_data['Industry'].isna().sum()
```

```python
stack_data['Industry'] = stack_data['Industry'].fillna('Unknown')
```

```python
stack_data['ConvertedCompYearly'] = stack_data['ConvertedCompYearly'].fillna(0)
```

```python
industry = stack_data[['Industry', 'ConvertedCompYearly']]
industry
```

```python
industry_avg = industry.groupby("Industry")["ConvertedCompYearly"].mean()
```

```python
highest_paying_industry = industry_avg.idxmax()
highest_paying_industry_val = industry_avg.max().round(2)
```

```python
industry_avg_dict = dict(industry_avg)
```

```python
industries = list(industry_avg_dict.keys())
compensation = list(industry_avg_dict.values())
```

```python
compensation
```

```python
industries
```

```python
plt.figure(figsize=(10, 6))
plt.barh(industries, compensation, color='skyblue')
plt.xlabel('Average Converted Compensation')
plt.ylabel('Industry')
plt.title('Average Converted Compensation by Industry')
plt.gca().invert_yaxis()
plt.tight_layout()

plt.show()
print(f"The highest paying industry for developers is {highest_paying_industry} with an average yearly salary of ${highest_paying_industry_val}")
```

### How much does remote working matter to employees?

I believe the best way to answer this question is to show how many remote workers there are in each age group and country to get a better understanding of the type of people who prefer/do remote work. As well as to see if location really does affect an employees decision to work remotely or not

```python
remote_work_df = stack_data[['RemoteWork', 'Country', 'Age']]
```

```python
remote_work_df = remote_work_df.dropna()
```

```python
grouped_by_country = remote_work_df.groupby(['Country'])['RemoteWork'].value_counts(normalize=True).unstack() * 100
```

```python
grouped_by_country.fillna(0)
```

```python
top_countries = grouped_by_country.index.get_level_values('Country').value_counts().head(15).index
grouped_by_country_subset = grouped_by_country.loc[top_countries]
```

```python
plt.figure(figsize=(12, 8))
grouped_by_country_subset.plot(kind='bar', stacked=True, colormap='coolwarm')
plt.title('Percentage of Remote Workers by Country')
plt.xlabel('Country')
plt.ylabel('Percentage')
plt.legend(title='Remote Work', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()

print(f"This visual shows the percentage of remote workers in the top 10 countries that developers reside in. Overall, we can see that most devlopers work remotely")
```

```python
excluded_values = ['Under 18 years old', 'Prefer not to say']
df_filtered = remote_work_df[~remote_work_df['Age'].isin(excluded_values)]

remote_age_distribution = df_filtered['Age'].value_counts()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(remote_age_distribution, labels=remote_age_distribution.index, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'lightcoral'])
plt.title('Distribution of Remote Work Preferences')
plt.axis('equal')
plt.show()

print(f"This pie chart shows that most remote workers are in the 25-34 years old range. This could be because people are starting families at this age and would prefer to be home.")
```

### How does coding experience affect the level of pay?

```python
years_coding['ConvertedCompYearly'] = years_coding['ConvertedCompYearly'].astype(int)
```

### What is the most popular form of learning to code?

```python
stack_data['LearnCode']
```

```python
stack_data.dropna(subset=['LearnCode'], inplace=True)
```

```python
separated_series = stack_data['LearnCode'].str.split(';').explode()
```

```python
value_counts = separated_series.value_counts()
```

```python
ways_of_learning = dict(value_counts)

methods = list(ways_of_learning.keys())
counts = list(ways_of_learning.values())
```

```python
plt.figure(figsize=(10, 6))
plt.barh(methods, counts, color='#85af31')
plt.xlabel('# of people')
plt.ylabel('Method')
plt.title('Most Popular forms of Learning to Code')
plt.gca().invert_yaxis()
plt.tight_layout()

plt.show()
print(f"According to this visual, the most popular form of learning code is online")
```
